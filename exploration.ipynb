```python
import pandas as pd
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect(r'C:\Users\BilalAsgharAziz\OneDrive - Powercloud GmbH\Documents\Data Engineering\2023-amse-template\project\data\cycles.sqlite')

# Retrieve the table names
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
table_names = cursor.fetchall()

# Read the tables into Pandas DataFrames
dataframes = {}  # Dictionary to store DataFrames
for table in table_names:
    table_name = table[0]
    query = f"SELECT * FROM {table_name};"
    df = pd.read_sql_query(query, conn)
    dataframes[table_name] = df

# Perform data analysis
for table_name, df in dataframes.items():
    print(f"Table: {table_name}")
    print(df.head())
    print('\n')

# Close the database connection
conn.close()
```
